# Argus

### 1. Wykorzystując bibliotekę scipy wygeneruj dwa zbiory danych w poniższych rozkładach:

#### a. https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.argus.html#scipy.stats.argus

```python
from scipy.stats import argus
import matplotlib.pyplot as plt
import seaborn as sns
chi = 1
data = argus.rvs(chi, size=1000)
print(data)
```

### 2. Wykorzystując bibliotekę matplotlib stwórz dla każdego zbioru:


#### a. wykresy skrzynkowe

```python
fig4, ax4 = plt.subplots()
ax4.set_title('Hide Outlier Points')
ax4.boxplot(data, showfliers=False)
plt.show()
```

#### b. histogramy

```python
plt.hist(data)
plt.title("Histogram")
plt.show()
```

### 3. Wykorzystując bibliotekę seaborn stwórz dla każdego zbioru:

#### a. wykresy histogramów

```python
sns.distplot(data)
plt.show()
```

#### b. wykresy jądrowych estymatorów gęstości

```python
sns.kdeplot(data, shade=True, color="b")
plt.show()
```

#### c. wykresy gęstościowe

```python
sns.distplot(data, kde=True, kde_kws={"color": "b", "shade":True })
plt.show()
```
